## Count frequency of numbers in dict
List me har number kitni baar repeat hua hai.

nums = [1, 2, 2, 3, 3, 3]

```python
nums = [1, 2, 2, 3, 3, 3]

freq = {}

for num in nums:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

## Find duplicate numbers
nums = [1, 2, 2, 3, 4, 4, 5]

```python
nums = [1, 2, 2, 3, 4, 4, 5]

freq = {}

for num in nums:
    if num in freq:
        freq[num]+= 1
    else:
        freq[num] = 1

duplicates = []

for num in freq:
    if freq[num] > 1:
        duplicates.append(num)

print(duplicates)
```

## Find maximum frequency number
nums = [1, 2, 2, 3, 3, 3, 4]


```python
nums = [1, 2, 2, 3, 3, 3, 4]

freq = {}

for num in nums:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1


max_freq = max(freq, key=freq.get)
print(max_freq)
```

## Find minimum frequency number
nums = [1, 2, 2, 3, 3, 3, 4]

```python
nums = [1, 2, 2, 3, 3, 3, 4]

freq = {}

for num in nums:
    if num in freq:
        freq[num]+= 1
    else:
        freq[num] = 1

minimum_freq = min(freq, key = freq.get)
print(minimum_freq)
```

## Merge two dictionaries
dict1 = {"a": 1, "b": 2}

dict2 = {"c": 3, "d": 4}

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}

merge_dict = dict1 | dict2

print(merge_dict)

```

## Sum all values in dictionary
marks = {"math": 80, "english": 75, "science": 90}

```python
marks = {"math": 80, "english": 75, "science": 90}

total = 0

for value in marks.values():
    total += value

print(total)
```

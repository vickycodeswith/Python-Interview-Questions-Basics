## Create a Rectangle class with length and breadth. Add methods to calculate area and perimeter.

```python
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def cal_area(self):
        area = self.length * self.breadth
        print(f"Area: {area}")

    def cal_perimeter(self):
        perimeter = 2 * (self.length + self.breadth)
        print(f"Perimeter: {perimeter}")


rect1 = Rectangle(10, 5)
rect1.cal_area()
rect1.cal_perimeter()
```

## Create a class Calculator with methods add, subtract, multiply, and divide.

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        result = self.num1 + self.num2
        print(f"Addition: {result}")

    def subtract(self):
        result = self.num1 - self.num2
        print(f"Subtraction: {result}")

    def multiply(self):
        result = self.num1 * self.num2
        print(f"Multiplication: {result}")

    def divide(self):
        if self.num2 != 0:
            result = self.num1 / self.num2
            print(f"Division: {result}")
        else:
            print("Cannot divide by zero")


calc1 = Calculator(10, 5)

calc1.add()
calc1.subtract()
calc1.multiply()
calc1.divide()
```

##  Create a multilevel inheritance example: GrandParent -> Parent -> Child.

```python
class GrandParent:
    def __init__(self, family_member):
        self.family_member = family_member

    def show_family(self):
        print(f"Family Name: {self.family_member}")

class Parent(GrandParent):
    def __init__(self, family_member, parent_name):
        super().__init__(family_member)
        self.parent_name = parent_name

    def show_parent(self):
        print(f"Parent Name: {self.parent_name}")


class child(Parent):
    def __init__(self, family_member, parent_name, child_name):
        super().__init__(family_member , parent_name)
        self.child_name = child_name

    def show_child(self):
        print(f"Child Name: {self.child_name}")


child1 = child("Yadav" , "Yadav g" , "Vicky")

child1.show_family()
child1.show_parent()
child1.show_child()
```

## Employee → Manager inheritance

```python
class Employee:
    def __init__(self, name, salary, department):
        self.name = name
        self.salary = salary
        self.department = department

    def display_employee(self):
        print(f"Employee Name: {self.name}")
        print(f"Salary: {self.salary}")
        print(f"Department: {self.department}")


class Manager(Employee):
    def __init__(self, name, salary, department, team_size):
        super().__init__(name, salary, department)
        self.team_size = team_size

    def display_manager(self):
        self.display_employee()
        print(f"Team Size: {self.team_size}")
        print("Manager can manage team")


manager1 = Manager("Rohit", 60000, "IT", 10)
manager1.display_manager()
```

### Create class Payment with method pay(). Create UPIPayment, CardPayment, and CashPayment.

```python
class Payment:
    def pay(self):
        print("Payment method")


class UPIPayment(Payment):
    def pay(self):
        print("Payment done using UPI")


class CardPayment(Payment):
    def pay(self):
        print("Payment done using Card")


class CashPayment(Payment):
    def pay(self):
        print("Payment done using Cash")


upi = UPIPayment()
card = CardPayment()
cash = CashPayment()

upi.pay()
card.pay()
cash.pay()
```

## Private Balance using Encapsulation
private variable banane ke liye variable ke aage double underscore __ lagate hain.

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.__balance = balance   

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited: {amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient Balance")

    def get_balance(self):
        return self.__balance


account1 = BankAccount("Vicky", 5000)

account1.deposit(2000)
account1.withdraw(1000)

print(f"Current Balance: {account1.get_balance()}")
```

## Abstraction
Create abstract class Payment with abstract method make_payment() sol


abstract class banane ke liye abc module use hota hai.

```python
from abc import ABC, abstractmethod


class Payment(ABC):
    @abstractmethod
    def make_payment(self):
        pass


class UPIPayment(Payment):
    def make_payment(self):
        print("Payment done using UPI")


class CardPayment(Payment):
    def make_payment(self):
        print("Payment done using Card")


class CashPayment(Payment):
    def make_payment(self):
        print("Payment done using Cash")


upi = UPIPayment()
card = CardPayment()
cash = CashPayment()

upi.make_payment()
card.make_payment()
cash.make_payment()
```

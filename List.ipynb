```python
text = "machine learning"

vowels = 'a' , 'e' , 'i', 'o' , 'u' 
count = 0
for char in text:
    if char in vowels:
       count += 1
print(count)
```

## Reverse a string
text = "python"

```python
text = "python"

reversed_text = text[:: -1]

print(reversed_text)


  # Without slicing
text = "python"

reversed_text = " "

for char in text:
    reversed_text = char + reversed_text

print(reversed_text)
```

slicing, [::-1] means start from the end of the string and move backwards one step at a time. So it reverses the complete string.”

## Count occurrence of an element

## Move zeroes to end
nums = [0, 1, 0, 3, 12]

```python
nums = [0, 1, 0, 3, 12]

new_nums = [ ]
zero_count = 0

for num in nums:
    if num != 0:
        new_nums.append(num)
    else:
        zero_count += 1

for i in range (zero_count):
    new_nums.append(0)

print(new_nums)

```

## Merge two lists
a = [1, 2, 3]

b = [4, 5, 6]

```python
a = [1, 2, 3]
b = [4, 5, 6]

merge = []

for num in a:
    merge.append(num)
for num in b:
    merge.append(num)
    
print(merge)
```

## List comprehension

nums = [1, 2, 3, 4]


```python
nums = [1, 2, 3, 4]
squares = [ num * num for num in nums ]

print(squares)



#. using For loop 
nums = [1, 2, 3, 4]

squares = []

for num in nums:
    squares.append(num*num)

print(squares)

```

## Find positive and negative numbers
nums = [-1, 2, -3, 4, 5]

## Find index of all occurrences
nums = [1, 2, 3, 2, 4, 2]
target = 2

 output: [1, 3, 5]

```python
nums = [1, 2, 3, 2, 4, 2]
target = 2

occurrence = []

for index, num in enumerate(nums):
    if num == target:
        occurrence.append(index)

print(occurrence)
```

## Find duplicate elements
nums = [1, 2, 2, 3, 4, 4, 5]

```python
nums = [1, 2, 2, 3, 4, 4, 5]

duplicate = []

for num in nums:
    if nums.count(num) > 1 and num not in duplicate:
        duplicate.append(num)

print(duplicate)

```

### Check Palindrome

```python
text = "madam"

reverse_text = text[::-1]

if text == reverse_text:
    print('palindrom')
else:
    print('Not palindrome')
```

## Count frequency of each character
text = "banana"

```python
text = "banana"

freq = { }

for char in text:
    if char in freq:
        freq[char]+= 1
    else:
        freq[char] = 1

print(freq)

```

## Find duplicate characters
text = "programming"


```python
text = "programming"

freq = {}

for char in text:
    freq[char] = freq.get(char, 0) + 1

duplicates = []

for char in freq:
    if freq[char] > 1:
        duplicates.append(char)

print(duplicates)
```

## Find longest word in a sentence
sentence = "I love machine learning"

```python
sentence = "I love machine learning"

words = sentence.split()
longest = ""
for word in words:
    if len(word) > len(longest):
        longest = word

print(longest)
```

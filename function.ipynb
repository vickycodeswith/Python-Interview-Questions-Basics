## Write a function to find the maximum number in a list.

```python
def max_number(nums):
    max_num = nums[0]

    for num in nums:
        if num > max_num:
            max_num = num
    return max_num

print(max_number([10, 20, 40, 30, 60, 80, 70]))
```

## Write a function to count vowels in a string.

```python
text = "machine learning"

def count_vowels(text):
    vowels = 'a' , 'e', 'i', 'o', 'u'
    count = 0

    for char in text:
        if char in vowels:
            count += 1
    return count

print(count_vowels(text))


```

## Write a function to check palindrome.

```python
def check_palindrome(text):
    if text == text[:: -1]:
        return "Palindrome"
    else:
        return "Not Palindrome"

text = "madam"

print(check_palindrome(text))
```

## Write a function to find all positions of a target element.

```python
def find_position(nums, target):
    
    position = []

    for i in range (len(nums)):
        if nums[i] == target:
            position.append(i)

    return position

print(find_position([1, 2, 3, 4, 5, 6, 7, 8, 9, 6], 6 ))
```

## Write a function to find duplicate elements.

```python
def duplicate_element(nums):
    duplicates = []

    for num in nums:
        if nums.count(num) > 1 and num not in duplicates:
            duplicates.append(num)
    return duplicates

nums = [1, 2, 2, 3, 4, 4, 5]
print(duplicate_element(nums))
```

## Write a function to calculate factorial.

```python
def factorial(n):
    result = 1

    for i in range(1, n + 1):
        result *= i

    return result

print(factorial(5))
```

## Write a recursive function for factorial.

Recursion means a function calling itself. Every recursive function must have a base condition to stop.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*factorial(n-1)

print(factorial(5))
```

## Write a function to return the second largest number.

```python
def second_largest_num(nums):
    unique_nums = list(set(nums))
    unique_nums.sort()

    return unique_nums[-2]

print(second_largest_num([10, 20, 30, 40 ,50, 30 , 20, 80 ]))
```

## Write a function to count frequency of each element.

```python
def freq(nums):
    freq = {}
    for num in nums:
        if num in freq:
            freq[num]+=1
        else:
            freq[num] = 1
    return freq

print(freq([10, 20, 3, 4, 6, 3, 5, 9, 10, 30, 50, 10]))
```

## Write a function to remove duplicates from a list.

```python
def remove_duplicates(nums):
    result= []
    for num in nums:
        if num not in  result:
            result.append(num)
    return result

print(remove_duplicates([10,10, 30, 40, 50, 30, 50, 60, 70, 90, 30]))
```
